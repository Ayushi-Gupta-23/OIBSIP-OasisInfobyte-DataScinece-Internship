### Training the model

### Logistic Regression can perform well with the linear dataset.
### It works by modeling the probability of each class, and if the decision boundaries between classes are mostly linear, Logistic Regression can provide accurate results.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression  # Example model

# Load your dataset (assuming 'df' contains your dataset)
# Replace 'features' and 'target' with the appropriate column names
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']

# Split the data into training and testing sets (e.g., 70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Data preprocessing: Standardize the feature values (optional but often beneficial)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Choose a machine learning algorithm (e.g., Logistic Regression as an example)
model = LogisticRegression(random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)

# Evaluate the model on the testing data
accuracy = model.score(X_test, y_test)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.xlabel("True Labels (y_test)")
plt.ylabel("Predicted Labels (y_pred)")
plt.title("True vs. Predicted Labels")
plt.show()
```

```python
# Create an array of indices for the data points
indices = np.arange(len(y_test))

# Create a line plot with separate y-axes
fig, ax1 = plt.subplots(figsize=(8, 6))
color = 'tab:blue'
ax1.set_xlabel("Data Point Index")
ax1.set_ylabel("True Labels", color=color)
ax1.plot(indices, y_test, marker='o', linestyle='-', color=color, label='True Labels')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Create a second y-axis that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel("Predicted Labels", color=color)
ax2.plot(indices, y_pred, marker='o', linestyle='-', color=color, label='Predicted Labels')
ax2.tick_params(axis='y', labelcolor=color)

plt.title("True vs. Predicted Labels (Line Graph with Separate Y-Axes)")
fig.tight_layout()
plt.show()
```

```python
# Visualize the model's predictions
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, c='blue', marker='o', alpha=0.7, label='Actual vs. Predicted')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', linewidth=2, label='Perfect Fit')
plt.xlabel("Actual ")
plt.ylabel("Predicted ")
plt.title("Actual vs. Predicted")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
